```python
import pandas as pd

saldo_inicial = 1000  # Saldo inicial
taxa_juros_anual = 0.12  # Taxa de juros anual (13%)

# Converta a taxa de juros anual para uma taxa di√°ria
taxa_juros_diaria = (1 + taxa_juros_anual) ** (1/365) - 1

# Crie um DataFrame com 30 dias
dias = list(range(1, 31))
saldos = []

for dia in dias:
    ganho_diario = saldo_inicial * (1 + taxa_juros_diaria) ** dia
    saldos.append(ganho_diario)

data = {'Dia': dias, 'Saldo': saldos}
df = pd.DataFrame(data)

display(df)
df.to_excel('diario.xlsx', index=False)
```
